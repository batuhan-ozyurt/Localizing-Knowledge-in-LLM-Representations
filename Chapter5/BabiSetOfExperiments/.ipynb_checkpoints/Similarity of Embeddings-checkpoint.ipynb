```python
from transformers import AutoTokenizer, AutoModelForSeq2SeqLM
from datasets import load_dataset
import transformers
import torch
import os
import nltk
import pandas as pd
import torch
import torch.nn as nn
import numpy as np
from jinja2 import Template
import pickle
from collections import defaultdict
import copy

from transformers.generation.logits_process import LogitsProcessorList
logits_processor = LogitsProcessorList()

import sys
sys.path.append('/scratch/users/bozyurt20/hpc_run/utilities')
sys.path.append("/scratch/users/bozyurt20/hpc_run/blobs/")
from util_research import *

max_len = 512
num_layers = 24
d_model = 4096

tokenizer = AutoTokenizer.from_pretrained("bigscience/T0pp", truncation_side="right", add_prefix_space=True)
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")

prng = np.random.RandomState(1)

babi_dataset_task1 = load_dataset('babi_qa', type='en-10k', task_no='qa1')
```

```python
model = AutoModelForSeq2SeqLM.from_pretrained("bigscience/T0pp", device_map='balanced')
#model.eval()
```

```python
class BabiDataPoint():
    def __init__(self, story, question, answer):
        self.story = story
        self.question = question
        self.answer = answer
    def add_story_encodings(self, encoding):
        self.story_encoding = encoding     
    def add_story_question_encodings(self, encoding):
        self.story_question_encoding = encoding

num_ex = len(babi_dataset_task1["train"])
all_data_points = []
uniqueness = []
for i in range(num_ex):
    story_text = babi_dataset_task1["train"][i]["story"]["text"]
    types = babi_dataset_task1["train"][i]["story"]["type"]
    
    story = ""
    for j, sent in enumerate(story_text):
        if types[j] == 0:
            story += sent + " "
        elif types[j] == 1:
            
            question = sent
            answer = babi_dataset_task1["train"][i]["story"]["answer"][j]
            data_point = BabiDataPoint(story[:-1], question, answer)
            
            if data_point.story + data_point.question not in uniqueness:
                uniqueness.append(data_point.story + data_point.question)
                all_data_points.append(data_point)

                
num_ex = len(babi_dataset_task1["test"])

for i in range(num_ex):
    story_text = babi_dataset_task1["test"][i]["story"]["text"]
    types = babi_dataset_task1["test"][i]["story"]["type"]
    
    story = ""
    for j, sent in enumerate(story_text):
        if types[j] == 0:
            story += sent + " "
        elif types[j] == 1:
            question = sent
            answer = babi_dataset_task1["test"][i]["story"]["answer"][j]
            data_point = BabiDataPoint(story[:-1], question, answer)
            
            if data_point.story + data_point.question not in uniqueness:
                uniqueness.append(data_point.story + data_point.question)
                all_data_points.append(data_point)
```

```python
## ENCODING THE CONTEXTS

for data_point in all_data_points:
    story = data_point.story
    question = data_point.question
    answer = data_point.answer
    
    input_ids = tokenizer.encode(story, return_tensors="pt")
    input_ids = input_ids.to(device)
    
    with torch.no_grad():
        encoder_out = model.encoder(input_ids)
        
    story_encoding = encoder_out.last_hidden_state.detach().cpu()
    data_point.add_story_encodings(story_encoding)
    
    context = story + " " + question
    
    input_ids = tokenizer.encode(context, return_tensors="pt")
    input_ids = input_ids.to(device)
    
    with torch.no_grad():
        encoder_out = model.encoder(input_ids)
    
    story_question_encoding = encoder_out.last_hidden_state.detach().cpu()
    data_point.add_story_question_encodings(story_question_encoding)
```

```python
def find_ind(story, entity): # both are strings.
    story_ids = tokenizer.encode(story)
    entity_id = tokenizer.encode(entity)[0]
    inds = []
    for i, story_id in enumerate(story_ids):
        if story_id == entity_id:
            inds.append(i)
    return inds
```

```python
def similarity_based_matching(with_question=False, similarity_function="cosine"):

    predictions = []
    
    for data_point in all_data_points:
        story = data_point.story
        question = data_point.question
        answer = data_point.answer
        
        if with_question:
            story_hiddens = data_point.story_question_encoding
        else:
            story_hiddens = data_point.story_encoding
        
        loc_inds = {}
        for loc in locs:
            inds = find_ind(story, loc)
            if len(inds) == 0:
                continue
            else:
                loc_inds[loc] = inds

        for char in chars:
            if char in question:
                query_char = char
                break

        query_char_ind = find_ind(story, query_char)[-1]

        loc_inds_values = list(loc_inds.values())

        all_loc_inds = []

        for inds in loc_inds_values:
            all_loc_inds += inds

        all_loc_inds = sorted(all_loc_inds)

        location_entity_encodings = story_hiddens[0, all_loc_inds, :] # num_locs, d
        character_entity_encoding = story_hiddens[0, query_char_ind, :] # d
        
        if similarity_function == "cosine":
            character_entity_encoding = character_entity_encoding.unsqueeze(0) # 1, d
            similarities = cosine_similarity(character_entity_encoding, location_entity_encodings) # num_locs
        elif similarity_function == "dot":
            similarities = torch.matmul(location_entity_encodings, character_entity_encoding) # num_locs
        elif similarity_function == "euclidian":
            similarities = torch.sum( (location_entity_encodings - character_entity_encoding)**2, dim=1) # num_locs

        assert similarities.shape[0] == len(all_loc_inds)

        loc_ind_position = torch.argmax(similarities).item()

        answer_ind = all_loc_inds[loc_ind_position]

        for loc, inds in loc_inds.items():
            if answer_ind in inds:
                answer_predicted = loc
                break
       
        if answer_predicted == answer:
            predictions.append(1)
        else:
            predictions.append(0) 

    accuracy = sum(predictions)/len(predictions)   
    
    return accuracy
```

```python
print(similarity_based_matching(with_question=False, similarity_function="cosine"))
print(similarity_based_matching(with_question=False, similarity_function="dot"))
print(similarity_based_matching(with_question=False, similarity_function="euclidian"))
```
