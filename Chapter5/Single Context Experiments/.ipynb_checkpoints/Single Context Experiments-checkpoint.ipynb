```python
template = tm8
one_sentence_probabilities = []
weird_cases = []
failure_cases = []
for sample in data_points_one_sentence:
    context = sample.context
    question = sample.question_1
    answer = sample.answer_1
    prompt = template.render(context=context, question=question)
    input_ids = tokenizer.encode(prompt, return_tensors="pt").to(device)
    model_output = model.generate(input_ids, max_new_tokens=1, output_scores=True, return_dict_in_generate=True)
    next_token_scores = torch.nn.functional.softmax(
                    model_output.scores[0], dim=-1
                )  # (batch_size * num_beams, vocab_size)
    
    probability = next_token_scores[0][tokenizer.encode(answer)[0]].item()
    scores = []
    for i, score in enumerate(next_token_scores[0]):
        scores.append( (i, score.item()) )

    scores.sort(key=lambda x: x[1], reverse=True)
    scores = [(tokenizer.decode(a), b) for a, b in scores]
    
    if tokenizer.decode(model_output.sequences[0], skip_special_tokens=True) != scores[0][0]:
        weird_cases.append(sample)
    if probability != scores[0][1]:
        failure_cases.append((sample))
    one_sentence_probabilities.append(probability)
    
```

```python
class DataSample_TwoSentences():
    def __init__(self, name_1, name_2, city_1, city_2):
        self.name_1 = name_1
        self.name_2 = name_2
        self.city_1 = city_1
        self.city_2 = city_2
        self.context = name_1 + " travelled to " + city_1 + ". " + name_2 + " travelled to " + city_2 + "."
        self.question_1 = "Where did " + name_1 + " travel to?"
        self.question_2 = "Where did " + name_2 + " travel to?"
        self.answer_1 = city_1
        self.answer_2 = city_2
        
class TwoSentenceResults():
    def __init__(self, prompt, context, question, probability_correct, probability_wrong):
        self.prompt = prompt
        self.context = context
        self.question = question
        self.probability_correct = probability_correct
        self.probability_wrong = probability_wrong
        
data_points_two_sentences = []
for name_1 in names_1:
    for name_2 in names_2:
        for city_1 in cities_1:
            for city_2 in cities_2:
                sample = DataSample_TwoSentences(name_1, name_2, city_1, city_2)
                data_points_two_sentences.append(sample)
```

```python
for sample in data_points_two_sentences:
    context = sample.context
    question_1 = sample.question_1
    question_2 = sample.question_2
    answer_1 = sample.answer_1
    answer_2 = sample.answer_2
    print(context, question_1, answer_1, question_2, answer_2)
```

```python
len(data_points_two_sentences)
```

```python
two_sentence_results = []
failure_cases = []
weird_cases = []
for sample in data_points_two_sentences:
    context = sample.context
    question_1 = sample.question_1
    question_2 = sample.question_2
    answer_1 = sample.answer_1
    answer_2 = sample.answer_2
    
    prompt_1 = template.render(context=context, question=question_1)
    input_ids = tokenizer.encode(prompt_1, return_tensors="pt").to(device)
    model_output = model.generate(input_ids, max_new_tokens=1, output_scores=True, return_dict_in_generate=True)
    next_token_scores = torch.nn.functional.softmax(
                    model_output.scores[0], dim=-1
                )  # (batch_size * num_beams, vocab_size)
    
    probability_correct = next_token_scores[0][tokenizer.encode(answer_1)[0]].item()
    probability_wrong = next_token_scores[0][tokenizer.encode(answer_2)[0]].item()
    
    scores = []
    for i, score in enumerate(next_token_scores[0]):
        scores.append( (i, score.item()) )

    scores.sort(key=lambda x: x[1], reverse=True)
    scores = [(tokenizer.decode(a), b) for a, b in scores]
    
    if tokenizer.decode(model_output.sequences[0], skip_special_tokens=True) != scores[0][0]:
        weird_cases.append((sample, question_1))
    if probability_correct != scores[0][1]:
        failure_cases.append((sample))
    
    result = TwoSentenceResults(prompt_1, context, question_1, probability_correct, probability_wrong)
    two_sentence_results.append(result)
    
    prompt_2 = template.render(context=context, question=question_2)
    input_ids = tokenizer.encode(prompt_2, return_tensors="pt").to(device)
    model_output = model.generate(input_ids, max_new_tokens=1, output_scores=True, return_dict_in_generate=True)
    next_token_scores = torch.nn.functional.softmax(
                    model_output.scores[0], dim=-1
                )  # (batch_size * num_beams, vocab_size)
    
    probability_correct = next_token_scores[0][tokenizer.encode(answer_2)[0]].item()
    probability_wrong = next_token_scores[0][tokenizer.encode(answer_1)[0]].item()
                             
    scores = []
    for i, score in enumerate(next_token_scores[0]):
        scores.append( (i, score.item()) )

    scores.sort(key=lambda x: x[1], reverse=True)
    scores = [(tokenizer.decode(a), b) for a, b in scores]
    
    if tokenizer.decode(model_output.sequences[0], skip_special_tokens=True) != scores[0][0]:
        weird_cases.append((sample, question_2))
    if probability_correct != scores[0][1]:
        failure_cases.append((sample))
    
    result = TwoSentenceResults(prompt_2, context, question_2, probability_correct, probability_wrong)
    two_sentence_results.append(result)
```

```python
class DataSample_OneSentence_Q2():
    def __init__(self, name_1, city_1):
        self.name_1 = name_1
        self.city_1 = city_1
        self.context = name_1 + " travelled to " + city_1 + "."
        self.question_1 = "Give a one word answer: What is the latest location of " + name_1 + " ?"
        self.answer_1 = city_1
        
data_points_one_char_two_loc = {}
for name_1 in names_1 + names_2:
    data_points_one_char_two_loc[name_1] = []
    for city_1 in cities_1 + cities_2:
        sample = DataSample_OneSentence_Q2(name_1, city_1)
        data_points_one_char_two_loc[name_1].append(sample)
```

```python
template = tm8
ti_two_sentence_probabilities_correct = []
ti_two_sentence_probabilities_wrong = []
weird_cases = []
failure_cases = []
for char in data_points_one_char_two_loc:
    for i in range(len(data_points_one_char_two_loc[char])-1):
        sample_1 = data_points_one_char_two_loc[char][i]
        sample_2 = data_points_one_char_two_loc[char][i+1]
        
        context_1 = sample_1.context
        question_1 = sample_1.question_1
        answer_1 = sample_1.answer_1
        
        context_2 = sample_2.context
        question_2 = sample_2.question_1
        answer_2 = sample_2.answer_1
        
        context = context_1 + " " + chapter2_part2_part1 + " Then, " + context_2 + " " + chapter2_part2_part2
        prompt = template.render(context=context, question=question_1)
        input_ids = tokenizer.encode(prompt, return_tensors="pt").to(device)
        model_output = model.generate(input_ids, max_new_tokens=1, output_scores=True, return_dict_in_generate=True)
        next_token_scores = torch.nn.functional.softmax(
                        model_output.scores[0], dim=-1
                    )  # (batch_size * num_beams, vocab_size)

        probability_correct = next_token_scores[0][tokenizer.encode(answer_2)[0]].item()
        probability_wrong = next_token_scores[0][tokenizer.encode(answer_1)[0]].item()
        
        scores = []
        for i, score in enumerate(next_token_scores[0]):
            scores.append( (i, score.item()) )

        scores.sort(key=lambda x: x[1], reverse=True)
        scores = [(tokenizer.decode(a), b) for a, b in scores]

        if tokenizer.decode(model_output.sequences[0], skip_special_tokens=True) != scores[0][0]:
            weird_cases.append((sample, question_1))
        if probability_correct != scores[0][1]:
            failure_cases.append((sample_1, sample_2, question_1, model_output))
            
        ti_two_sentence_probabilities_correct.append(probability_correct)
        ti_two_sentence_probabilities_wrong.append(probability_wrong)
```
