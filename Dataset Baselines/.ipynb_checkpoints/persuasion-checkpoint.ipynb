```python
import re

import nltk
import pandas as pd
import torch
import numpy as np
import xmltodict
import pickle
from collections import defaultdict

from fuzzywuzzy import fuzz

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize

import sys
sys.path.append('/scratch/users/bozyurt20/hpc_run/utilities')
from util_research import *
```

```python
path_persuasion = "/kuacc/users/bozyurt20/hpc_run/105_persuasion.txt"
path_annotations = "/kuacc/users/bozyurt20/hpc_run/PersuasionWork/Persuasion Annotations"

with open(path_persuasion, "r") as f:
    story = f.read()

story = story.lower()

dir_list_annotations = os.listdir(path_annotations)
all_annotations = {}

for item in dir_list_annotations:
    
    f = open(os.path.join(path_annotations, item), 'r')
    annotations = pd.read_csv(f, sep="\t")
    annotations = annotations.values
    f.close()
    
    all_annotations[item] = annotations

characters = []
locations = []

for item in all_annotations:
    
    annotations = all_annotations[item]
    
    for line in annotations:

        char = line[1]
        loc = line[2]

        if type(char) == float:
            pass
        else:
            chars = char.lower().split("/")
            characters.extend(chars)

        locs = loc.lower().split("/")
        locations.extend(locs)

set_chars = list(set(characters))
set_locs = list(set(locations))

#set_locs = set_locs[2:]
location_inds = defaultdict(list)

for loc in set_locs:
    location_inds[loc] = [m.start() for m in re.finditer(loc, story)]
    
location_inds2 = defaultdict(list)

for loc in set_locs:
    location_inds2[loc] = [m.start() for m in re.finditer('(?='+loc+')', story)]
    
assert location_inds2 == location_inds

all_loc_ind_tuples = []
all_loc_inds = []
for loc_name, loc_inds in location_inds.items():
    for loc_ind in loc_inds:
        all_loc_ind_tuples.append((loc_name, loc_ind))
        all_loc_inds.append(loc_ind)

loc_dict = {}
for tupl in all_loc_ind_tuples:
    loc_name, loc_ind = tupl
    loc_dict[loc_ind] = loc_name 
    
all_loc_inds = np.array(all_loc_inds)

exact_accuracy = []
fuzzy_accuracy = []
tt = 0
cant_find = []
for item in all_annotations:
    
    annotations = all_annotations[item]
    
    for i, line in enumerate(annotations):
        
        ind = line[0]
        char_line = line[1].lower()
        loc = line[2].lower()
        
        chars = char_line.lower().split("/")
        locs = loc.lower().split("/")
        
        book_prev = story[:ind+1]
        
        char_inds = {}
        for char in chars:
            if char[:4] == "the ":
                char = char[4:]
            char_inds[char] = book_prev.rfind(char)

        temp_distance2loc = {}
        #print(char_inds)
        for char, char_ind in char_inds.items():
            if char_ind != -1:
                distances = []
                for a_loc_ind in all_loc_inds:
                    if a_loc_ind - char_ind > 0:
                        dist = a_loc_ind - (char_ind + len(char))
                    else:
                        dist = char_ind - a_loc_ind
                    distances.append(dist)
                distances = np.array(distances)
                #distances = np.abs(all_loc_inds - char_ind)
                distance = np.min(distances)
                x = np.argmin(distances)
                temp_loc_ind = all_loc_inds[x]
                temp_distance2loc[distance] = loc_dict[temp_loc_ind]
            else:
                pass
                #print(char)
                #print(i , item, ind)
        
        if len(temp_distance2loc) == 0:
            tt += 1
            cant_find.append(char_line)
            continue
        else:
            min_distance = min(list(temp_distance2loc.keys()))
            loc_prediction = temp_distance2loc[min_distance]
            match1, match2 = exactly_or_fuzzily_matched(loc_prediction, chars[0], locs)

            if match1 == "Yes":
                exact_accuracy.append(1)
            else:
                exact_accuracy.append(0)
            if match2 == "Yes":
                fuzzy_accuracy.append(1)
            else:
                fuzzy_accuracy.append(0)
        
print(sum(exact_accuracy)/len(exact_accuracy))
print(sum(fuzzy_accuracy)/len(fuzzy_accuracy))  
```
